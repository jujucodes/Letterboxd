```python
# add genre column to each dataframe
df_horror['genre'] = 'horror'
df_horror.head()
```

```python
df_scifi['genre'] = 'scifi'
df_scifi.head()
```

```python
df_thriller['genre'] = 'thriller'
df_thriller.head()
```

```python
df_war['genre'] = 'war'
df_war.head()
```

```python
df_animation['genre'] = 'animation'
df_animation.head()
```

```python
df_genres = df_horror.append(df_scifi)
df_genres.info()
```

```python
df_genres = df_genres.append(df_thriller)
df_genres = df_genres.append(df_war)
df_genres = df_genres.append(df_animation)
df_genres.info()
```

```python
df_genres.head()
```

```python
df_genres['genre'].value_counts()
```

```python
plt.hist(df_genres['genre'])
plt.show()
```

```python
plt.bar(df_genres['genre'],df_genres['avg_rating'])
plt.show()
```

```python
df_genres.isna().sum()
```

```python
df_genres['year_isna'] = df_genres['year'].isna()
df_genres['year_isna'].sum()
```

```python
df_genres.isna().sum()
```

```python
df_genres['year'] = df_genres['year'].fillna(df_genres['year'].mean())
df_genres.isna().sum()
```

```python
df_genres['director'] = df_genres['director'].fillna('directorNA')
df_genres.isna().sum()
```

```python
model = LinearRegression()
```

```python
X = pd.get_dummies(df_genres['genre'])
X
```

```python
y = df_genres['avg_rating']
model.fit(X,y)
```

```python
# predictive power for this feature
model.coef_
```

```python
y_pred = model.predict(X)
pd.Series(y_pred).unique()
```

```python
y_error = y_pred - y
y_error.mean()
```

```python
standard_error = ((y_error**2).mean())**0.5
standard_error
```
