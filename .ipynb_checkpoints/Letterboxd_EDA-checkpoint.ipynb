Add genre column to each DataFrame

```python
df_horror['genre'] = 'horror'
df_horror.head()
```

```python
df_scifi['genre'] = 'scifi'
df_scifi.head()
```

```python
df_thriller['genre'] = 'thriller'
df_thriller.head()
```

```python
df_war['genre'] = 'war'
df_war.head()
```

```python
df_animation['genre'] = 'animation'
df_animation.head()
```

Append DataFrames

```python
df_genres = df_horror.append(df_scifi)
df_genres.info()
```

```python
df_genres = df_genres.append(df_thriller)
df_genres = df_genres.append(df_war)
df_genres = df_genres.append(df_animation)
df_genres.info()
```

```python
df_genres.to_csv('df_genres.csv')
```

Inspect outliers

```python
print(df_genres[df_genres['avg_rating'] == df_genres['avg_rating'].min()])
```

Even the lowest rated films had a good number of five star ratings, indicating that these films may instead be more "controversial"

```python
print(df_genres[df_genres['avg_rating'] == df_genres['avg_rating'].max()])
```

Distribution of ratings are more gradual and linear for the highest rated films

Conclusions: Average ratings feature has a normal distribution, but genres have some important feature differences between them. Strong correlations between average ratings and the number of stars, however this is not useful as including these would cause data leakage and inhibit real predictions.
