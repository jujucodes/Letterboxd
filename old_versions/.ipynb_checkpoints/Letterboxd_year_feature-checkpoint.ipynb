```python
#data cleaning
```

```python
df_genres.isna().sum()
```

```python
# new columns with missing year and director data
```

```python
# df_genres['year_isna'] = df_genres['year'].isna()
# df_genres['year_isna'].sum()
df_genres['year'] = df_genres['year'].fillna(df_genres['year'].mean())
```

```python
df_genres['director'] = df_genres['director'].fillna('directorNA')
df_genres.isna().sum()
```

```python
# genre feature model training
```

```python
model = LinearRegression()
```

```python
X = pd.get_dummies(df_genres['genre'])
X
```

```python
y = df_genres['avg_rating']
```

```python
#name these better, add true to true data
X_train, X_test, y_train_true, y_test_true = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
model.fit(X_train, y_train_true)
```

```python
y_test_pred = model.predict(X_test)
```

```python
model.score(X_test, y_test_true)
```

```python
# predictive power for this feature
model.coef_
```

```python
pd.Series(y_test_pred).unique()
```

```python
rmse = np.sqrt(mean_squared_error(y_test_true, y_test_pred))
rmse
```

```python
#
y_error = y_test_pred - y_test_true
y_error
```

```python
standard_error = ((y_error**2).mean())**0.5
standard_error
```

```python
#compute z score
```

```python
# model with cross validation
```

```python
cv_scores = cross_val_score(model, X, y, cv=5)
np.mean(cv_scores)
```
