```python
feat_imp = gb_reg.feature_importances_
columns = X_train_true.columns
gbGraph = pd.Series(feat_imp, columns)
print(gbGraph)
gbGraph.sort_values().plot.barh(color ='blue')
plt.title('Gradient Boosting feature importance')
plt.savefig('leakage_gb_features.png')
```

Gradient Boosting Regressor and KNeighbors Regressor provide the most accurate predictions on unseen data.
